## 2.1 Αναγνωριστικά (Identifiers)

Ένα **αναγνωριστικό (identifier)** της Python είναι ένα _όνομα_ που χρησιμοποιείται για να αναγνωρίζονται τα εξής πράγματα (που θα ορισθούν/συζητηθούν κατά την διάρκεια των μαθημάτων):
* μια μεταβλητή
* μια συνάρτηση
* μια κλάση
* μια μονάδα (module)
* ή ένα αντικείμενο

Ένα αναγνωριστικό γράφεται με τα εξής _συνεχόμενα_ σύμβολα:
* τα λατινικά γράμματα, κεφαλαία ή μικρά,
* τον χαρακτήρα υπογράμμισης (underscore)
* τους αριθμούς από 0 ως 9

Κάθε αναγνωριστικό πρέπει:
* να περιλαμβάνει τουλάχιστον ένα (ή περισσότερα) από τα προηγούμενα σύμβολα (αλλά, αν αποτελείται μόνο από ένα σύμβολο, αυτό **δεν** πρέπει να είναι αριθμός!)
* και, σε οποιαδήποτε περίπτωση, **να μην αρχίζει με αριθμό**

## 2.3 Μπλοκ (γραμμών κώδικα) και Εσοχές (Indentation)

Ο κώδικας (οι εντολές) στην Python γράφονται σε διακριτά μπλοκ γραμμών. Όταν ένα μπλοκ περιέχει κάποιο άλλο μπλοκ, το δεύτερο πρέπει να γράφεται με εσοχή (indentation) από αριστερά. Σε κάθε διαδοχικό περιεχόμενο μπλοκ προστίθεται η αντίστοιχη εσοχή. Καθώς ο κώδικας εκτελείται, όπως είναι γραμμένος από πάνω (αριστερά) προς τα κάτω (δεξιά), δυο (ή περισσότερα) μπλοκ που το ένα ακολουθεί το άλλο (χωρίς το δεύτερο να περιέχεται στο δεύτερο) γράφονται με τις ίδιες εσοχές (ή και καθόλου, αν αυτά τα μπλοκ δεν περιέχονται σε κανένα άλλο).

Ένα παράδειγμα δομής μπλοκ-εσοχών εδίνεται από το παρακάτω απόσπασμα κώδικα (snippet), που αποτελεί έναν βρόχο (loop) for (όπως θα δούμε στη συνέχεια του μαθήματος):

```python
for i in [1, 2, 3, 4, 5]:
    print i                      # first line in "for i" block
    for j in [1, 2, 3, 4, 5]:
        print j                  # first line in "for j" block
        print i + j              # last line in "for j" block
    print i                      # last line in "for i" block
print "done looping"
```
